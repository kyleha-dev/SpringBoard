```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## 2 Load The E-commerce Data <a id='2_Load_The_E_commerce_Data'></a>

```python
order_data = pd.read_csv('../raw_data/ecommerce_data.csv')
```

```python
order_data.info()
```

note: 
- Row Id is the primary key for now
- Postal Code should be object (categorical)
- Order Date and Ship date should be dateTime

```python
del order_data["Row ID"]
order_data['Order Date'] = pd.to_datetime(order_data['Order Date'])
order_data['Ship Date'] = pd.to_datetime(order_data['Ship Date'])
order_data['Postal Code'] = order_data.astype(object)
order_data.info()
```

```python
order_data.head()
```

Table is joined from 3 different tables, Order, Customer and Product. Let's figure out how many orders, customers and products there were.

```python
for colName in order_data.columns:
    print('\n', '--------------------------------')
    print(order_data[colName].describe())
    print('--------------------------------','\n')
```

### 2.5.2 Number Of Missing Values By Column<a id='2.6.2_Number_Of_Missing_Values_By_Column'></a>

```python
missing = pd.concat([order_data.isnull().sum(), 100 * order_data.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing.sort_values(by='count')
```

#### 2.5.3 Distribution Of Orders By Region And State<a id='2.5.3_Distribution_Of_Resorts_By_Region_And_State'></a>

#### 2.5.4 Distribution Of Products By Category And Subcategory<a id='2.5.4_Distribution Of Orders By Category And Subcategory'></a>

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 8))

order_data['Category'].value_counts().plot(kind='barh', ax=ax[0])
ax[0].set_title('Category')
ax[0].set_xlabel('Count')

order_data['Sub-Category'].value_counts().plot(kind='barh', ax=ax[1])
ax[1].set_title('Sub-Category')
ax[1].set_xlabel('count')

plt.subplots_adjust(wspace=0.5);
```

```python
order_data.hist(figsize=(15,10))
plt.subplots_adjust(hspace=0.5);
```

```python
sub_category_profits = pd.melt(order_data[['Sub-Category', 'Profit']], 
                        id_vars='Sub-Category',
                        value_vars='Profit', 
                        value_name='Profit')
sub_category_profits.head()

plt.subplots(figsize=(12, 8))
sns.boxplot(x='Sub-Category', y='Profit', hue='variable', data=sub_category_profits)
plt.xticks(rotation='vertical')
plt.ylabel('Price ($)')
plt.xlabel('State');
```

```python
sub_category_profits = pd.melt(order_data[['Sub-Category', 'Sales']], 
                        id_vars='Sub-Category',
                        value_vars='Sales', 
                        value_name='Sales')
sub_category_profits.head()

plt.subplots(figsize=(12, 8))
sns.boxplot(x='Sub-Category', y='Sales', hue='variable', data=sub_category_profits)
plt.xticks(rotation='vertical')
plt.ylabel('Sales')
plt.xlabel('Sub-Category');
```

```python
sub_category_profits = pd.melt(order_data[['Sub-Category', 'Discount']], 
                        id_vars='Sub-Category',
                        value_vars='Discount', 
                        value_name='Discount')
sub_category_profits.head()

plt.subplots(figsize=(12, 8))
sns.boxplot(x='Sub-Category', y='Discount', hue='variable', data=sub_category_profits)
plt.xticks(rotation='vertical')
plt.ylabel('Discount')
plt.xlabel('Sub-Category');
```
